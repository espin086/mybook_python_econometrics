```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set(style="white", palette="muted", color_codes=True)
```

## Importing Raw Data

```python
def import_data(xls_name):
    df = pd.read_excel("../../data/{0}".format(xls_name), index_col=None, header=None)
    return df

def rename_df(df):
    df.rename(columns = {0:'salary', 
                     1:'pcsalary', 
                     2:'sales',
                     3:'roe', 
                     4:'pcroe', 
                     5:'ros',
                     6:'indus', 
                     7:'finance', 
                     8:'consprod',
                     9:'utility', 
                     10:'lsalary', 
                     11:'lsales',
                    },inplace = True)
    return df


def import_raw_data(xls_name):
    df = import_data(xls_name='ceosal1.xls')
    df = rename_df(df=df)
    return df

df_raw = import_raw_data(xls_name = 'ceosal1.xls')

!cat ../../data/CEOSAL1.DES

```
